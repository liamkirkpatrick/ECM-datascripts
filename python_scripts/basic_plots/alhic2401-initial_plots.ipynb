```python
#%% Import packages 

# general
import numpy as np
import pandas as pd

# plotting
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# my functions/classes
import sys
sys.path.append("../core_scripts/")
from ECMclass import ECM
```

```python
# smoothing window, mm
window = 10

# paths
path_to_data = '../../data/'
path_to_figures = '../../../figures/'
metadata_file = 'metadata.csv'
```

```python
#%% Read in metadata and import data

meta = pd.read_csv(path_to_data+metadata_file)

# import each script as an ECM class item
data = []
cores = []
sections = []
faces = []
ACorDCs = []
for index,row in meta.iterrows():

    core = row['core']
    section = row['section']
    face = row['face']
    ACorDC = row['ACorDC']
    
    if core == 'alhic2401':# and section == '40':
        
        print("Reading "+core+", section "+section+'-'+face+'-'+ACorDC)
    
        data_item = ECM(core,section,face,ACorDC)
        data_item.rem_ends(10)
        data_item.smooth(window)
        data_item.norm_all()
        #data_item.norm_outside()
        data.append(data_item)
        
        cores.append(core)
        sections.append(section)
        faces.append(face)
        ACorDCs.append(ACorDC)

sec = set(sections)
```

```python
def plotquarter(y_vec,ycor,d,meas,button,axs,rescale,cmap,plot_button = False):

    res = 0.002
    
    width = y_vec[1] - y_vec[0]
    
    for y in y_vec:
        
        
        idx = ycor==y
        
        tmeas = meas[idx]
        tbut = button[idx]
        tycor = ycor[idx]
        td = d[idx]

        #downsample ECM to save plotting time (as needed)
        if res != 0:
            int_lo = round(min(td),2)
            int_hi = round(max(td),2)
            depth_interp = np.linspace(int_lo,int_hi,int((int_hi-int_lo)/res)+1)
            meas_interp = np.interp(depth_interp,np.flip(td),np.flip(tmeas))
            but_interp = np.interp(depth_interp,np.flip(td),np.flip(tbut))
            td = depth_interp
            tmeas = meas_interp
            tbut = np.round(but_interp)
        
        for i in range(len(tmeas)-1):
            
            if tbut[i] == 0 or not plot_button:
                
                axs.add_patch(Rectangle((y-(width-0.2)/2,td[i]),(width-0.2),td[i+1]-td[i],facecolor=cmap(rescale(tmeas[i]))))
                #axs.add_patch(Rectangle((y-(width-0.2)/2,td[i]),(width-0.2),td[i+1]-td[i],facecolor='r'))
            else:
                axs.add_patch(Rectangle((y-(width-0.2)/2,td[i]),(width-0.2),td[i+1]-td[i],facecolor='k'))
            
    return()
```

```python
def unique(list1):
 
    # initialize a null list
    unique_list = []
 
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    
    return(unique_list)
```

```python
# make colormap
my_cmap = matplotlib.colormaps['Spectral']
```

```python
for plt_but in [True]:

    if plt_but:
        print('Plotting with buttons')
    else:
        print('Plotting without buttons')

    for sec in unique(sections):

        
        # print update
        print("    Running Section "+sec)
        
        # set data to empty
        AC_w1 = None
        AC_w2 = None
        DC_w1 = None
        DC_w2 = None


        #loop through data 
        for d in data:
            
            # find faces
            if d.core=='alhic2401':
                if d.section==sec:
                    if d.ACorDC == 'AC':
                        if d.face == 'w1':
                            AC_w1 = d
                        if d.face == 'w2':
                            AC_w2 = d
      
                    else:

                        if d.face == 'w1':
                            DC_w1 = d
                        if d.face == 'w2':
                            DC_w2 = d


        
        # find depth max and minimum
        minvec = []
        maxvec = []
        AC_all = []
        DC_all = []
        for data_face in [AC_w1,AC_w2,DC_w1,DC_w2]:
            if data_face != None:
                
                minvec.append(min(data_face.depth))
                maxvec.append(max(data_face.depth))
                
                mask = data_face.button == 0
                #mask2 = data_face.meas != nan
                #mask = mask1 * mask2
                if data_face.ACorDC == 'AC':
                    AC_all.extend(data_face.meas[mask])
                else:
                    DC_all.extend(data_face.meas[mask])
        ACpltmin = np.percentile(AC_all,5)
        ACpltmax = np.percentile(AC_all,95)
        DCpltmin = np.percentile(DC_all,5)
        DCpltmax = np.percentile(DC_all,95)  
        ACrescale = lambda k: (k-ACpltmin) /  (ACpltmax-ACpltmin)
        DCrescale = lambda k: (k-DCpltmin) /  (DCpltmax-DCpltmin)
        dmin = min(minvec)
        dmax = max(maxvec)
        
        # make figure
        fig, ax = plt.subplots(1, 5, gridspec_kw={'width_ratios': [3, 3,3, 3, 3]},figsize=(9,6),dpi=100)
        
        # right-specific
        for a in [ax[0],ax[3]]:
            #a.yaxis.tick_right()
            a.set_xlim([55, 0])
            
        # top specific
        for a in [ax[1],ax[4]]:
            a.yaxis.tick_right()
            a.yaxis.set_label_position("right")
            
            a.set_xlim([0,55])
            
        # applies to all
        for a in [ax[0],ax[1],ax[3],ax[4]]:
            a.set_ylabel('Depth (m)')
            a.set_xlabel('Distance From Center (mm)',fontsize=6)
            a.set_ylim([dmax, dmin])
            
        for a,data_face in zip([ax[0],ax[1],ax[3],ax[4]],[AC_w2,AC_w1,DC_w2,DC_w1]):
            
            if data_face != None:
                if data_face.face == 'w2':
                    yall = data_face.y_s - data_face.y_left
                    yvec = data_face.y_vec - data_face.y_left
                else:
                    yall = data_face.y_right - data_face.y_s
                    yvec = data_face.y_right - data_face.y_vec
                
                if data_face.ACorDC =='AC':
                    rescale = ACrescale
                else:
                    rescale = DCrescale
            
                # plot data
                plotquarter(yvec,
                            yall,
                            data_face.depth_s,
                            data_face.meas_s,
                            data_face.button_s,
                            a,
                            rescale,
                            my_cmap,
                            plot_button = plt_but)
        
        # housekeeping
        fig.suptitle('ALHIC2401 - '+sec+' - '+str(window)+' mm smooth')
        ax[2].axis('off')
        ax[0].set_title('AC - W2')
        ax[1].set_title('AC - W1')
        ax[3].set_title('DC - W2')
        ax[4].set_title('DC - W1')
        
        fig.tight_layout()
        plt.subplots_adjust(wspace=0)
    
        # ad colorbar
        #fig.subplots_adjust(bottom=0.8)
        #    ACcbar_ax = fig.add_axes([0.08,-0.07,0.35,0.05])
        ACcbar_ax = fig.add_axes([0.07,-0.05,0.35,0.05])
        ACnorm = matplotlib.colors.Normalize(vmin=ACpltmin,vmax=ACpltmax)
        DCcbar_ax = fig.add_axes([0.58,-0.05,0.35,0.05])
        DCnorm = matplotlib.colors.Normalize(vmin=DCpltmin,vmax=DCpltmax)
        ACcbar = fig.colorbar(matplotlib.cm.ScalarMappable(norm=ACnorm, cmap=my_cmap),cax=ACcbar_ax,
                    orientation='horizontal',label='Current (amps)')
        DCcbar = fig.colorbar(matplotlib.cm.ScalarMappable(norm=DCnorm, cmap=my_cmap),cax=DCcbar_ax,
                    orientation='horizontal',label='Current (amps)')
        
        # save figure
        if int(sec) <10:
            sec = '0'+sec
        if plt_but:
            fname = path_to_figures +'alhic2401_button/alhic2401-'+sec+'-button.png'
        else:
            fname = path_to_figures +'alhic2401_nobutton/alhic2401-'+sec+'-no-button.png'
        fig.savefig(fname,bbox_inches='tight')

        # fig.close()
```
